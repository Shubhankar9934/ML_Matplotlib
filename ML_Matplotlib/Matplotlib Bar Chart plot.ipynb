# Plotting Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
# Dataset of 'Indian Artificail Intelligence Production (IAIP) Class'

# classes = ['Python', 'R', 'AI','ML','DS']

classes = ["Python", "R", "AI", "ML", "DS"]
class1_students =[30,10,20,25,10] # out of 100 student in each classs
class2_students =[40,5,20,20,10]
class3_students =[35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
```

```python
# To plot bar char horizontally we use barh() method.
plt.barh(classes,class1_students)
```

```python
# To change the size of  bar chart()

plt.bar(classes,class1_students, width = 0.4)
```

```python
# To change the alignmaent of bar chart
plt.bar(classes,class1_students, width = 0.4, align = 'edge')
```

```python
# To change the color of bar chart we use color parameter
plt.bar(classes,class1_students, width = 0.4 , align = 'edge', color = 'g')
```

```python
# to change the color of edge of a bar.
plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r")
```

```python
# to change the line width of a bar
plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5 )
```

```python
# To change the intencity of a color we use alpha parameter.
plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5, alpha = 0.5)
```

```python
# to change the linestyle of a bar chart we use linestyle parameter.
plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5, alpha = 0.5, linestyle = '--')
```

```python
# to labelling the bar chart we use label parameter
plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5, alpha = 0.5, linestyle = '--', 
        label = "Class 1 Students")
plt.legend(loc =4)
```

```python
# to hide this barchart we use visible parameter
plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5, alpha = 0.5, linestyle = '--', 
        label = "Class 1 Students", visible = False)
plt.legend(loc =4)
```

```python
# To change the style of barplot we use style.use("ggplot")
style.use("ggplot")

plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5, alpha = 0.5, linestyle = '--', 
        label = "Class 1 Students", visible = True)
plt.legend(loc =4)
```

```python
# to change the size of bar chart we use plt.figure() method

style.use("ggplot")

plt.figure(figsize=[16,9])

plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5, alpha = 0.5, linestyle = '--', 
        label = "Class 1 Students", visible = True)
plt.legend(loc =4)
```

```python
# To plot bar chart Horizontally we use barh method and 
# remove multiple width else we got an error.
style.use("ggplot")

plt.figure(figsize=[16,9])

plt.barh(classes,class1_students , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5, alpha = 0.5, linestyle = '--', 
        label = "Class 1 Students", visible = True)
plt.legend(loc =4)
```

```python
# To add an title in a barchart we use plt.title() Parameter
style.use("ggplot")

plt.figure(figsize=[16,9])

plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5, alpha = 0.5, linestyle = '--', 
        label = "Class 1 Students", visible = True)
plt.legend(loc =4)
plt.title("Bar Chart of IAIP Class")
```

```python
# to add label in barchart we use xlable and y lable
# Also changing the font size of title and xlabel,ylabel.
style.use("ggplot")

plt.figure(figsize=[16,9])

plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5, alpha = 0.5, linestyle = '--', 
        label = "Class 1 Students", visible = True)
plt.legend(loc =4)
plt.title("Bar Chart of IAIP Class", fontsize = 30 , color = 'y')

plt.xlabel("Classes", fontsize =25)
plt.ylabel("No. of Students", fontsize = 25)

plt.show()
```

```python
# Plotting Multiple Bars in same barchart.
style.use("ggplot")

plt.figure(figsize=[16,9])

plt.bar(classes,class1_students, width = 0.4 , 
        align = 'edge', color = 'g',edgecolor = "r",
       linewidth = 5, alpha = 0.5, linestyle = '--', 
        label = "Class 1 Students", visible = True)

plt.bar(classes,class2_students, width = 0.4 , 
        align = 'edge', color = 'b',edgecolor = "m",
       linewidth = 5, alpha = 0.5, linestyle = '--', 
        label = "Class 2 Students", visible = True)

plt.legend(loc =4)
plt.title("Bar Chart of IAIP Class", fontsize = 30 , color = 'y')

plt.xlabel("Classes", fontsize =25)
plt.ylabel("No. of Students", fontsize = 25)

plt.show()
```

```python
# Formatting Barchart so that it doesn't overlap to each other.
style.use("ggplot")

plt.figure(figsize=[16,9])

classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index,class1_students, width, 
        align = 'edge', color = 'g',
        label = "Class 1 Students")

plt.bar(classes_index+width,class2_students, width, 
        align = 'edge', color = 'b',
        label = "Class 2 Students")

plt.title("Bar Chart of IAIP Class", fontsize = 30 , color = 'y')

plt.xlabel("Classes", fontsize =25)
plt.ylabel("No. of Students", fontsize = 25)

plt.show()
```

```python
# To show the value of classes we use plt.xticks() parameter
style.use("ggplot")

plt.figure(figsize=[16,9])

classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index,class1_students, width, 
        align = 'edge', color = 'g',
        label = "Class 1 Students")

plt.bar(classes_index+width,class2_students, width, 
        align = 'edge', color = 'b',
        label = "Class 2 Students")

plt.xticks(classes_index+width,classes,rotation = 30)

plt.title("Bar Chart of IAIP Class", fontsize = 30 , color = 'y')

plt.xlabel("Classes", fontsize =25)
plt.ylabel("No. of Students", fontsize = 25)

plt.show()
```

```python
# plotting Classes_3 also in same barchart.
style.use("ggplot")

plt.figure(figsize=[16,9])

classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index,class1_students, width, 
        align = 'edge', color = 'g',
        label = "Class 1 Students")

plt.bar(classes_index+width,class2_students, width, 
        align = 'edge', color = 'b',
        label = "Class 2 Students")

plt.bar(classes_index+width,class3_students, width, 
        align = 'edge', color = 'r',
        label = "Class 3 Students")

plt.xticks(classes_index+width,classes,rotation = 30)

plt.title("Bar Chart of IAIP Class", fontsize = 30 , color = 'y')

plt.xlabel("Classes", fontsize =25)
plt.ylabel("No. of Students", fontsize = 25)

plt.show()
```

```python
#Formatting Barchart again because bar chart is overlapping.
style.use("ggplot")

plt.figure(figsize=[16,9])

classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index,class1_students, width, 
        align = 'edge', color = 'g',
        label = "Class 1 Students")

plt.bar(classes_index+width,class2_students, width, 
        align = 'edge', color = 'b',
        label = "Class 2 Students")

plt.bar(classes_index+width+width,class3_students, width, 
        align = 'edge', color = 'r',
        label = "Class 3 Students")

plt.xticks(classes_index+width,classes,rotation = 30)

plt.title("Bar Chart of IAIP Class", fontsize = 30 , color = 'y')

plt.xlabel("Classes", fontsize =25)
plt.ylabel("No. of Students", fontsize = 25)

plt.show()
```

```python
# Plotting Horizontal bar.
style.use("ggplot")

plt.figure(figsize=[16,9])

classes_index = np.arange(len(classes))

width = 0.2

plt.barh(classes_index,class1_students, width, 
        align = 'edge', color = 'g',
        label = "Class 1 Students")

plt.barh(classes_index+width,class2_students, width, 
        align = 'edge', color = 'b',
        label = "Class 2 Students")

plt.barh(classes_index+width+width,class3_students, width, 
        align = 'edge', color = 'r',
        label = "Class 3 Students")

plt.yticks(classes_index+width,classes,rotation = 30)

plt.title("Bar Chart of IAIP Class", fontsize = 30 , color = 'y')

plt.ylabel("Classes", fontsize =25)
plt.xlabel("No. of Students", fontsize = 25)
plt.legend()
plt.show()
```
