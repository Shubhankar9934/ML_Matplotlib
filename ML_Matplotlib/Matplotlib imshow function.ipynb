# how to show image using matplotlib imshow Function?

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Reading of image 
img = mpimg.imread("pi_chart.png")
img
```

```python
# Cheaking type of image
type(img)
```

```python
# cheaking the shape of an image 
img.shape
```

```python
#cheaking the dimension of an image
img.ndim
```

```python
# showing of an image 
plt.imshow(img)
plt.show()
```

```python
# removing of an axis of an image 
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
# increasing the size ofan image 
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
# showing the color bar of an image 
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
#Assing the value of color map using the cmap param
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img,cmap="hot")
plt.colorbar()
plt.show()
```

```python
# given data of single channel to cmap so that it works perfectally.
single_cahnnel = img[:,:,1]
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(single_cahnnel,cmap="hot") # cmap means color map
plt.colorbar()
plt.show()
```
