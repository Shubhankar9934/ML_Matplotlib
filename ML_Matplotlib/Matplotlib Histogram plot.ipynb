# Ploting Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age = np.random.randint(18,45,(100)) 
# creating Random 100 variable
py_students_age = np.random.randint(15,40,(100))
```

```python
print(ml_students_age) 
print(py_students_age)
```

```python
plt.hist(ml_students_age) # Printing histogram
plt.show()
```

```python
# Including Title
plt.hist(ml_students_age) # Printing histogram
plt.title("ML Student age Histogram")
plt.show()
```

```python
# Alloting level on x axis and y axis.
plt.hist(ml_students_age)
plt.title("ML Student age Histogram")

plt.xlabel("Student age Category")
plt.ylabel("No. Student age")

plt.show()
```

# bins


```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins)
plt.title("ML Student age Histogram")

plt.xlabel("Student age Category")
plt.ylabel("No. Student age")

plt.show()
```

# rwidth

```python
plt.hist(ml_students_age, bins, rwidth=0.8)
plt.title("ML Student age Histogram")

plt.xlabel("Student age Category")
plt.ylabel("No. Student age")

plt.show()
```

# histtype

```python
plt.hist(ml_students_age, bins, rwidth=0.8, histtype = 'step')
plt.title("ML Student age Histogram")

plt.xlabel("Student age Category")
plt.ylabel("No. Student age")

plt.show()
```

# orientation

```python
plt.hist(ml_students_age, bins, rwidth=0.8, histtype = 'bar',
         orientation = 'horizontal')
plt.title("ML Student age Histogram")

plt.xlabel("Student age Category")
plt.ylabel("No. Student age")

plt.show()
```

# color

```python
plt.hist(ml_students_age, bins, rwidth=0.8, histtype = 'bar',
         orientation = 'horizontal',color ='r')
plt.title("ML Student age Histogram")

plt.xlabel("Student age Category")
plt.ylabel("No. Student age")

plt.show()
```

# label

```python
plt.hist(ml_students_age, bins, rwidth=0.8, histtype = 'bar',
         orientation = 'horizontal',color ='r',label = 'Ml student')
plt.title("ML Student age Histogram")

plt.xlabel("Student age Category")
plt.ylabel("No. Student age")

plt.legend(loc =4)

plt.show()
```

```python
# Ploting two value in same histogram
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8, histtype = 'bar',
         orientation = 'vertical',color ='r',label = 'Ml student')

plt.hist(py_students_age, bins, rwidth=0.8, histtype = 'bar',
         orientation = 'vertical',color ='y',label = 'py student')

plt.title("ML Student age Histogram")

plt.xlabel("Student age Category")
plt.ylabel("No. Student age")

plt.legend(loc =4)

plt.show()
```

# Proper formating of histogram

```python
# Ploting two value in same histogram
bins = [15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age],bins, rwidth=0.8, histtype = 'bar',
         orientation = 'vertical',color =["m","y"],label = ['Ml student','py student'])

# plt.hist(py_students_age, bins, rwidth=0.8, histtype = 'bar',
#          orientation = 'vertical',color ='y',label = 'py student')

plt.title("ML Student age Histogram")

plt.xlabel("Student age Category")
plt.ylabel("No. Student age")

plt.legend(loc =4)

plt.show()
```

# figure size


```python
# Ploting two value in same histogram
bins = [15,20,25,30,35,40,45]

plt.figure(figsize = (16,9))

plt.hist([ml_students_age,py_students_age],bins, rwidth=0.8, histtype = 'bar',
         orientation = 'vertical',color =["m","y"],label = ['Ml student','py student'])

# plt.hist(py_students_age, bins, rwidth=0.8, histtype = 'bar',
#          orientation = 'vertical',color ='y',label = 'py student')

plt.title("ML Student age Histogram")

plt.xlabel("Student age Category")
plt.ylabel("No. Student age")

plt.legend(loc =4)

plt.show()
```
