```python
# Changing Color in Scatter Plot
plt.scatter(x,y,color = "r")
plt.title(" Google Play Store Scatter Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
# To increase the Size of Scatter Plot we use plt.figure() parameter.
plt.figure(figsize=(16,9))
plt.scatter(x,y,color = "r")
plt.title(" Google Play Store Scatter Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
# To change the marker in scatter plot we use  marker param 
plt.figure(figsize=(16,9))
plt.scatter(x,y,color = "r", marker = "*")
plt.title(" Google Play Store Scatter Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
# Now Marker is changed into Dot(.) to astricks(*)
```

```python
# To change the size of marker 
plt.figure(figsize=(16,9))
plt.scatter(x,y,color = "r", marker = '*', s = 100) #markersize = s
plt.title(" Google Play Store Scatter Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
# To cahnge the transparency of scatter marker we use alpha param
plt.figure(figsize=(16,9))
plt.scatter(x,y,color = "r", marker = '*', s = 100 , alpha = 0.2) #markersize = s
plt.title(" Google Play Store Scatter Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
# To change the line width of marker we use linewidth param
plt.figure(figsize=(16,9))
plt.scatter(x,y,color = "r", marker = '*', s = 100, linewidths = 10) #markersize = s
plt.title(" Google Play Store Scatter Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
# To change edge color of scatter plot we use edgecolor param
plt.figure(figsize=(16,9))

plt.scatter(x,y,color = "r", marker = '*', s = 100,
        linewidths = 10, edgecolors = 'g') #markersize = s

plt.title(" Google Play Store Scatter Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
 #Plotting Multiple Scatter in same Scatter Graph
plt.figure(figsize=(16,9))

plt.scatter(x,y,color = "r", marker = '*', s = 100,
        linewidths = 10, edgecolors = 'g') #markersize = s

plt.scatter(x,df_google_play_store_apps["Installs"],color = "y", marker = 'o', s = 100,
        linewidths = 10, edgecolors = 'c')

plt.title(" Google Play Store Scatter Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews Nd Install")
plt.show()
```
